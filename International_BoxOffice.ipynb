```python
def title_boxoffice():
    y = {}
    for url in url_list:
        page = requests.get(url)
        soup = BeautifulSoup(page.content,'html.parser')
        page = soup.find('div',class_="a-section imdb-scroll-table-inner")
        for tr in page.findAll('tr'): 
            x = tr.find('a').get_text()
            if x == 'Rank':
                continue
            else:
                key = x
                val = (tr.find('td',class_='a-text-right mojo-field-type-money')).contents
                y[key]=val
    return y 
k = title_boxoffice() 
k
```
