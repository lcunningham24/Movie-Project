```python
merge1 = pd.merge(basics[['tconst','primary_title','runtime_minutes']], ratings[['tconst','averagerating']], right_on=None)
merge1
```

```python
gross_genre_df = pd.read_csv('IMDB_scraped.csv')
```

```python
gross_genre_df.head()
```

```python
# renaming the columns 
merge1['Movie Title'] = merge1['primary_title']
merge1['IMDB Rating'] = merge1['averagerating']
merge1['Runtime (min)'] = merge1['runtime_minutes']
```

```python
merge1
```

```python
merge2 = pd.merge(gross_genre_df[['Movie Title','Gross Revenue (USD)','Genre']],merge1[['tconst','Movie Title','Runtime (min)','IMDB Rating']],on='Movie Title')
merge2.head(50)
```

```python
merge2.isna().sum()
```

```python
merge3 = merge2.drop_duplicates(subset='Movie Title',keep='first')
merge3
```

```python
merged_data = merge3.to_csv('Merged_IMDB_Data.csv')
```
