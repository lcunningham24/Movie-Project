```python
with gzip.open('./data/imdb.title.basics.csv.gz') as f:

    basics= pd.read_csv(f)

basics.shape
```

```python
with gzip.open('./data/imdb.title.ratings.csv.gz') as f:

    ratings= pd.read_csv(f)

ratings
```

```python
with gzip.open('./data/tn.movie_budgets.csv.gz') as f:

    budget = pd.read_csv(f)

budget.head(10)
```

```python
merge1 = pd.merge(basics, ratings[['tconst','averagerating']], right_on=None)
merge1
```

```python
gross_genre_df = pd.read_csv('IMDB_scraped.csv')
```

```python
gross_genre_df.head()
```

```python
# renaming the columns 
merge1['Movie Title'] = merge1['primary_title']
merge1['IMDB Rating'] = merge1['averagerating']
merge1['Runtime (min)'] = merge1['runtime_minutes']
```

```python
merge1.head()
```

```python
merge2 = pd.merge(gross_genre_df[['Movie Title','Gross Revenue (USD)','Genre']],merge1[['Movie Title','Runtime (min)','IMDB Rating']],on='Movie Title')
merge2
```

```python
merge3 = merge2.drop_duplicates(subset='Movie Title',keep='first')
merge3.head()
```

```python
merged_data = merge3.to_csv('Merged IMDB Data.csv')
```
