```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
# list of IMDB urls to webscrape 

num_list = [str(i) for i in np.arange(51,4952,50)]

def url():
    '''function that returns list of urls to webscrape'''
    url_list = ['https://www.imdb.com/search/title/?release_date=2010-01-01,2020-01-01&sort=boxoffice_gross_us,desc&view=advanced']
    for num in num_list: 
        url ='https://www.imdb.com/search/title/?release_date=2010-01-01,2020-01-01&sort=boxoffice_gross_us,desc&start='+num+'&ref_adv_nxt'
        url_list.append(url)
    return url_list 
url_list = url()
```

```python
# list of titles 
def movie_titles():
    title_list = []
    for url in url_list: 
        page = requests.get(url)
        soup = BeautifulSoup(page.content,'html.parser')
        page = soup.find('div',class_="lister-list")
        
        titles = [h3.find('a').get_text() for h3 in page.findAll('h3')]
        title_list.append(titles)
    return title_list
```

```python
l = movie_titles()
```

```python
# turning list of lists into one list 
titles = []
for sublist in l:
    for item in sublist:
        titles.append(item)
titles
```

```python
len(titles)
```

```python
# list of revenues 
revenue_list = []
for url in url_list: 
    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')
    page = soup.find('div',class_="lister-list")
    for a in page.findAll('p',class_="sort-num_votes-visible"):
        x = a.findAll('span')[-1]
        revenue_list.append(x['data-value'])
revenue_list
```

```python
len(revenue_list)
```

```python
# list of genres
genre_list = []
for url in url_list: 
    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')
    page = soup.find('div',class_="lister-list")
    for genre in page.findAll('p',class_="text-muted"):
        genres = []
        genre = genre.find('span',class_="genre")
        genres.append(genre)
        for genre in genres: 
            if genre != None: 
                genre_list.append(genre.text.strip('\n').rstrip())
            else: 
                continue 
```

```python
genre_list
```

```python
# list of IMDB ratings 
ratings = []
for url in url_list: 
    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')
    page = soup.find('div',class_="lister-list")
    for rating in page.findAll('div',class_="inline-block ratings-imdb-rating"):
        ratings.append(rating.attrs["data-value"])
ratings
```

```python
import pandas as pd 
df1 = pd.DataFrame([titles, revenue_list, genre_list]).transpose()
df1.columns = ['Movie Title','Gross Revenue ($)','Genre']
df1
```
